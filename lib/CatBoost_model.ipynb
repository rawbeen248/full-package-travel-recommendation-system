```python
def record_interaction(userid, city, cat, price,price_airbnb,beds_airbnb,people_airbnb,reviews_airbnb,place_category, place_time, time, indicator,pool_df):
        #global pool_df
        new_row = pd.DataFrame({col: [0] for col in pool_df.columns})
        new_row[shared_cols] = user_df.loc[(user_df['UserId']==userid)][shared_cols].values
        time = datetime.strptime(time, '%Y-%m-%d %H:%M:%S')
        new_row['inptTime'] = time
        new_row['inptCat'] = cat
        new_row['inptPrice'] = price
        new_row['inptCity'] = city
        new_row['inptPriceAir'] = price_airbnb
        new_row['inptBedsAir'] = beds_airbnb
        new_row['inptPeopleAir'] = people_airbnb
        new_row['inptReviewsAir'] = reviews_airbnb
        new_row['inptPlaceCat'] = place_category
        new_row['inptPlaceTime'] = place_time
        if indicator == 0:
            new_row['Label'] = 0
        else:
            new_row['Label'] = 1
        pool_df = pool_df.append(new_row, ignore_index=True)
        return pool_df
```

```python
def gen_random_date():
    random_datetime = start + timedelta(seconds=random.randint(0, delta.total_seconds()))
    random_datetime_str = random_datetime.strftime("%Y-%m-%d %H:%M:%S")
    return random_datetime_str
```

```python
#ENCODING
from sklearn.preprocessing import OrdinalEncoder

cat_cols = ['inptCity', 'inptCat', 'inptPlaceCat','inptPlaceTime']

# Create an instance of the encoder
encoder = OrdinalEncoder()

# Fit the encoder to the categorical columns
encoder.fit(pool_df[cat_cols])

# Transform the categorical columns using the encoder
pool_df[cat_cols] = encoder.transform(pool_df[cat_cols])
pool_df = pool_df.drop(['inptTime'], axis=1)

pool_df
```

```python
from sklearn.model_selection import train_test_split #split train test, with set random state for reproducible results
X_train = pool_df.drop(['Label'], axis=1)
y_train = pool_df['Label']

X_train, X_test, y_train, y_test = train_test_split(X_train, y_train, test_size=0.3, random_state=42)
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score
from sklearn.metrics import classification_report

def make_report(y_true, y_pred): #function to quickly print metrics for multi-label classifier
    target_names = ['class 0', 'class 1']
    print(classification_report(y_true, y_pred, target_names=target_names))
    
from catboost import CatBoostClassifier
from sklearn.metrics import accuracy_score
from sklearn import metrics
```

```python
ctb = CatBoostClassifier(bootstrap_type='Bernoulli', boosting_type='Plain', 
                         iterations=1500, loss_function='Logloss', depth=6
                        , leaf_estimation_method='Gradient',boost_from_average=True,
                        score_function='Cosine') #i love catboost so i also used it
ctb.fit(X_train, y_train)
y_pred_ctb = ctb.predict(X_test)
probs = ctb.predict_proba(X_test)
probs = [arr[1] for arr in probs]
new_threshold = 0.2
'''y_pred_ctb_thresh = np.where(probs[:, 1] > new_threshold, 1, 0)'''
# To try and reduce false positives I played around with the decision threshold but found no value that performed better than the standard


ctb_accuracy = accuracy_score(y_test, y_pred_ctb)
print('Catboost Accuracy: ', ctb_accuracy)
print(metrics.confusion_matrix(y_test, y_pred_ctb))
print(metrics.roc_auc_score(y_test, probs))
make_report(y_test, y_pred_ctb)
```
